```python
from nltk.corpus import stopwords
import re
import nltk
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def clean_tweet(tweet):
    tweet=re.sub(r'[^A-Za-z0-9\s]','',tweet)
    tweet_tokens = tweet.split()
    filtered_tweet = ' '.join([word for word in tweet_tokens if word.lower() not in stop_words])
    return filtered_tweet
```

```python
df
```

```python
df['Cleaned Tweet Content'] = df['Tweet Content'].apply(clean_tweet)
```

```python
df
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
def analyze_sentiment(tweet):
    sentiment = analyzer.polarity_scores(tweet)
    return sentiment['compound']
```

```python
df['Sentiment Score'] = df['Tweet Content'].apply(analyze_sentiment)
```

```python
def categorize_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df['Sentiment Category'] = df['Sentiment Score'].apply(categorize_sentiment)
```
